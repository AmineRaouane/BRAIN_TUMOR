```python
import pandas as pd  # Pandas for reading files
import numpy as np  # Numpy for math

import seaborn as sns
import matplotlib.pyplot as plt # Seaborn,Plotly and matplotlib for plotting and visualization
plt.style.use('default') # Set the global style
```

```python
Data_Brain_Tumor = pd.read_csv("Brain Tumor.csv")
Data_Brain_Tumor.replace([np.inf, -np.inf], np.nan, inplace=True)
Data_Brain_Tumor.drop("Image", axis=1, inplace=True)
```

```python
Data_Brain_Tumor.describe()
```

```python
Data_Brain_Tumor.head()
```

```python
Data_Brain_Tumor.info() # Get the information of the data
```

```python
from sklearn.model_selection import train_test_split


Train , Test = train_test_split(Data_Brain_Tumor, test_size=0.2, random_state=42,stratify=Data_Brain_Tumor['Class'])
```

```python
import xgboost as xgb
```

```python
model = xgb.XGBClassifier()
```

```python
model.fit(Train.drop("Class", axis=1),Train['Class'])
```

```python
predictions = model.predict(Test.drop("Class", axis=1))
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(Test['Class'],predictions)
```

```python
import joblib

joblib.dump(model, "XGB.pkl")
```
